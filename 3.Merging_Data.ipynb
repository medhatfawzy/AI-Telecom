Time to clean and process the Traffic Volume Dataset

```python
Traffic_Data_Grouped = sqldf(
    """SELECT DISTINCT Timestamp, LocationLatitude, LocationLongitude, RadioOperatorName, RadioNetworkGeneration, RadioConnectionType, RadioMobileDataEnabled, TrafficDirection, SUM(TrafficVolume) AS TrafficVolume
    FROM Traffic_volume_data 
    GROUP BY Timestamp, 
            LocationLatitude,
            LocationLongitude, 
            RadioOperatorName,
            TrafficDirection,
            RadioNetworkGeneration,
            RadioConnectionType,
            RadioMobileDataEnabled
    ORDER BY Timestamp, 
            LocationLatitude,
            LocationLongitude, 
            RadioOperatorName,
            TrafficDirection,
            RadioNetworkGeneration, 
            RadioMobileDataEnabled
    ;
""", globals())
```

Here I grouped the data by summing the traffic volume

```python
Traffic_Data_Grouped.info()
```

Our dataframe columns have default types again we need to convert them back

```python
cols = [
            "RadioOperatorName",
            "TrafficDirection",
            "RadioNetworkGeneration",
            "RadioConnectionType",
            "RadioMobileDataEnabled"
      ]

Traffic_Data_Grouped[cols] = Traffic_Data_Grouped[cols].astype('category')
```

```python
Traffic_Data_Grouped["Timestamp"] = pd.to_datetime(Traffic_Data_Grouped["Timestamp"])
```

```python
Traffic_Data_Grouped.info()
```

Now all types are back the same

```python
Traffic_Data_Grouped.head()
```

Now let's join these two dataframe. I'll first join them based on the ["LocationLongitude", "LocationLatitude", "RadioOperatorName", "RadioNetworkGeneration", "RadioConnectionType", "RadioMobileDataEnabled"] Columns. Then I will start joining based on the timestamp column.

```python
merged_data = pd.merge(RSRP_data,
                       Traffic_Data_Grouped,
                       how='inner',
                       on=["LocationLongitude", 
                            "LocationLatitude",
                            "RadioOperatorName", 
                            "RadioNetworkGeneration",
                            "RadioConnectionType", 
                            "RadioMobileDataEnabled"
                           ]
                       )
```

```python
merged_data.info()
```

By looking at the dataset we can see that the RSRP value is obtained at random periods, while the Traffic Volume value is obtained every 15 minutes. This is I'll be joining on the same day, hour but not the same minute. When it comes to the minutes I'll join only the RSRP minute is within the 15 min range of the Traffic Volume.

```python
merged_data = merged_data[
                            (merged_data["Timestamp_x"].dt.date ==  merged_data["Timestamp_y"].dt.date)
                               &
                            (merged_data["Timestamp_x"].dt.hour ==  merged_data["Timestamp_y"].dt.hour)
                               &
                            (merged_data["Timestamp_x"].dt.minute >= merged_data["Timestamp_y"].dt.minute)
                               &
                            (merged_data["Timestamp_y"].dt.minute > merged_data["Timestamp_x"].dt.minute -15)
                           ]
```

```python
merged_data.info()
```

```python
merged_data[cols] = merged_data[cols].astype('category')
```

```python
merged_data.info()
```

```python
merged_data.head(20)
```

```python
merged_data.sort_values(by="Timestamp_x", inplace=True)
merged_data.reset_index(drop=True, inplace=True)
```

```python
merged_data.head(20)
```

Now I'll drop the Timestamp_x column and group the RSRP using the average, this will give us a nice and clean time-series dataset

```python
merged_data.drop("Timestamp_x", axis=1, inplace=True)
```

```python
merged_data.head(10)
```

```python
merged_data_grouped = sqldf(
    """SELECT DISTINCT Timestamp_y AS Timestamp,
                        LocationLatitude, 
                        LocationLongitude, 
                        RadioOperatorName, 
                        RadioNetworkGeneration, 
                        RadioConnectionType, 
                        RadioMobileDataEnabled, 
                        DeviceManufacturer, 
                        DeviceName, 
                        TrafficDirection, 
                        TrafficVolume,
                        AVG(RSRP) AS RSRP
        FROM merged_data 
        GROUP BY Timestamp_y,
                LocationLatitude, 
                LocationLongitude, 
                RadioOperatorName, 
                RadioNetworkGeneration, 
                RadioConnectionType, 
                RadioMobileDataEnabled, 
                DeviceManufacturer, 
                DeviceName, 
                TrafficDirection, 
                TrafficVolume
        ORDER BY Timestamp, 
                LocationLatitude,
                LocationLongitude, 
                RadioOperatorName,
                TrafficDirection,
                RadioNetworkGeneration, 
                RadioMobileDataEnabled
;
""", globals())
```

```python
merged_data_grouped.info()
```

We can see that we have multiple RSRP values for every 15 min that's why we see a huge reduction is the size of our dataset

```python
merged_data_grouped.head(10)
```

```python
Down_traffic = merged_data_grouped[merged_data_grouped["TrafficDirection"] == "Downlink"].copy()
Up_traffic = merged_data_grouped[merged_data_grouped["TrafficDirection"] == "Uplink"].copy()
```

```python
Down_traffic.reset_index(drop=True, inplace=True)
Up_traffic.reset_index(drop=True, inplace=True)
```

```python
cols_all = cols + ["DeviceManufacturer", "DeviceName"]

Down_traffic[cols_all] = Down_traffic[cols_all].astype('category')
Down_traffic["Timestamp"] = pd.to_datetime(Down_traffic["Timestamp"])

Up_traffic[cols_all] = Up_traffic[cols_all].astype('category')
Up_traffic["Timestamp"] = pd.to_datetime(Up_traffic["Timestamp"])

```
